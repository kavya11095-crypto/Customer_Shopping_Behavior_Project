```python
# Loading the dataset using pandas
import pandas as pd
df=pd.read_csv('downloads/customer-trends-data-analysis-SQL-Python-PowerBI-main/customer-trends-data-analysis-SQL-Python-PowerBI-main/customer_shopping_behavior.csv')
```

```python
# here we are finding the missing values 
df.isnull().sum()
```

```python
df.['review rating'] = df.groupby('category')['review rating'].transform(lambda x: x.fillna(x.median()))
```

```python
df.isnull().sum()
```

```python
#here we are making the column hedar in profer way lower case and replacing space with _
df.columns = df.columns.str.lower()
df.columns = df.columns.str.replace(' ','_')
df=df.rename(columns={'Customer ID':'customer_id'})
df=df.rename(columns={'purchase_amount_(usd)':'purchase_amount'})
```

```python
df.columns
```

```python
# create a column age_group
labels = ['young adult', 'adult', 'middle-aged', 'senior']
df['age_group'] = pd.qcut(df['age'], q=4, labels=labels)
```

```python
df[['age','age_group']].head(10)
```

```python
 # create new column purchase_frequency_days

frequency_mapping = {
    'Fortnightly': 14,
    'Weekly': 7,
    'Monthly': 30,
    'Quarterly': 90,
    'Bi-Weekly': 14,
    'Annually': 365,
    'Every 3 Months': 90
}

df['purchase_frequency_days'] = df['frequency_of_purchases'].map(frequency_mapping)
```
